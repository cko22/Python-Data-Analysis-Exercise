$$\min\limits_{B} ~ \mathbb{E}(E^TE)$$

$$ Let ~ M = \mathbb{E}[(Y- X B^T)^T(Y - X B^T)]$$

$$ = \mathbb{E}[(Y^TY -2B^TX^TY + BX^TXB^T)]$$

Find $ \hat{B}^T$ by $\frac{\partial M}{\partial B} = 0$

$$ \frac{\partial M}{\partial B} = \frac{\partial \mathbb{E}[(Y^TY -2B^TX^TY + BX^TXB^T)]}{\partial B}$$

$$0  = -2X^TY+2X^TXB^T $$

$$ \Rightarrow \hat{B}^T = (X^TX)^{-1}(X^TY)$$

<h2> Question 3 </h2>

```python
# Defining Variables
Y = data[0:168,0:13]
Tick = df.columns[0:13] #Tickers' name
a = np.ones(168) # Create vector with elements of 1s.
X = np.empty((168,2))
X[:,0] = a
X[:,1] = data[0:168,13]
```

<h3> Question 3.1 </h3>

```python
B_hat = np.matmul(np.matmul(np.transpose(Y),X),np.linalg.inv(np.matmul(np.transpose(X),X)))
table3_1 ={'Alpha': B_hat[:,0], 'Beta': B_hat[:,1]}
pd.DataFrame(data=table3_1)
```

<h3> Question 3.2 </h3>

The sample covariance matrix of $ \hat{E} = Y - \hat{B}^TX$ can be calculated by the following covariance formula,

$$ Cov(\epsilon_k,\epsilon_j) = \frac{\Sigma_{i=1}^{n}(\epsilon_{ik} - \bar{\epsilon_{k}})(\epsilon_{ij} - \bar{\epsilon_{j}})}{n-2} $$

where $k, j = 1, 2, ..., 13$ and $n = 168$. This formula calculates one element in the covariance matrix, to obtain the full matrix for loop is used. The code is shown below. 

```python
E_hat = Y - np.matmul(X,np.transpose(B_hat))
Sam_cov = []
Vec_cov = []
for k in range(13): # For each row
    for j in range(13): # For each column
        Cov = 0
        for i in range(168): # For each k-j element,
            Cov += (E_hat[i,k] - np.mean(E_hat[:,k]))*(E_hat[i,j] - np.mean(E_hat[:,j]))/(168-2)
        Vec_cov.append(Cov) # Collect the covariances as a vector
Sam_cov = np.reshape(Vec_cov,(13,13))   # Reshape into matrix form 
table3_2 = pd.DataFrame(Sam_cov)
table3_2.style.set_caption("Covariance Matrix of E_hat")
```

<h3> Question 3.3 </h3>

Using the given formula, $$ \sigma_i^2 = [Cov(\hat{E})]_{i,i} $$

```python
Var = []
for i in range(13):
    Var.append(Sam_cov[i,i]) # Get the diagonal elements of sample covariance matrix.
table3_3 = {'Variance': Var}
pd.DataFrame(table3_3)
```

<h3> Question 3.4 </h3>

Using the given formula, $$ R_i^2 = 1 - \frac{[\hat{E}^T\hat{E}]_{i,i}}{[Y^TY]_{i,i}} $$

```python
R_2 = []
for i in range(13):
    R_2.append(1 - (np.matmul(np.transpose(E_hat[:,i]),E_hat[:,i]))/(np.matmul(np.transpose(Y[:,i]),Y[:,i])))
table3_4 = {'R-squared':R_2}
pd.DataFrame(table3_4)
```

<h3> Question 3.5 </h3>

```python
beta_hat = B_hat[:,1].tolist()   
sigma = np.sqrt(Var).tolist()  
table_2 ={'beta': beta_hat,'sigma': sigma,'R-squared': R_2}
pd.DataFrame(data=table_2,index =Tick)
```

<h3> Question 6.1 </h3>

$$\min\limits_\stackrel{\rightarrow}{\omega} ~ \stackrel{\rightarrow}{\omega}^{T}\Sigma\stackrel{\rightarrow}{\omega}$$

$$subject~to~\stackrel{\rightarrow}{\omega}^{T}\mathbf{1} = 1~,$$

Applying the Lagrangian Function, 

$$\mathscr{L} = \stackrel{\rightarrow}{\omega}^{T}\Sigma\stackrel{\rightarrow}{\omega} - \lambda(\stackrel{\rightarrow}{\omega}^{T}\mathbf{1} - 1)~.$$

Compute the gradient,

$$
\nabla_{\stackrel{\rightarrow}{\omega},\lambda}\mathscr{L} = 
\begin{bmatrix}
  2\stackrel{\rightarrow}{\omega}^{T}\Sigma - \lambda\mathbf{1}^T \\
  1 - \stackrel{\rightarrow}{\omega}^{T}\mathbf{1}   
\end{bmatrix}
= \begin{bmatrix}
  0 \\
  0   
\end{bmatrix}
$$ 

Solve the linear system, 

$$2\stackrel{\rightarrow}{\omega}^{T}\Sigma - \lambda\mathbf{1}^T = 0 \tag{1}$$

$$1 - \stackrel{\rightarrow}{\omega}^{T}\mathbf{1} =0 \tag{2}$$

From Equation $(1)$,

$$2\stackrel{\rightarrow}{\omega}^{T}\Sigma = \lambda\mathbf{1}^T$$

$$\stackrel{\rightarrow}{\omega}^{T}\Sigma\Sigma^{-1} = \frac{\lambda}{2}\mathbf{1}^T\Sigma^{-1}$$

$$\Rightarrow \stackrel{\rightarrow}{\omega}^{T} = \frac{\lambda}{2}\mathbf{1}^T\Sigma^{-1} \tag{3}$$

From Equation $(2)$,

$$\stackrel{\rightarrow}{\omega}^{T}\mathbf{1} = 1$$

Substituting into Equation $(3)$,

$$\frac{\lambda}{2}\mathbf{1}^T\Sigma^{-1}\mathbf{1} = 1$$

$$\Rightarrow \lambda = \frac{2}{\mathbf{1}^T\Sigma^{-1}\mathbf{1}}$$

Now solve for $\stackrel{\rightarrow}{\omega}$,

$$\stackrel{\rightarrow}{\omega}=(\stackrel{\rightarrow}{\omega}^{T})^{T}$$

$$\stackrel{\rightarrow}{\omega}=(\frac{\lambda}{2}\mathbf{1}^T\Sigma^{-1})^T$$

Using the expression for $\lambda$ gives,

$$\stackrel{\rightarrow}{\omega}=\frac{\big(\frac{2}{\mathbf{1}^T\Sigma^{-1}\mathbf{1}}\big)}{2}{\Sigma^{-1}}^T\mathbf{1}$$

$$\stackrel{\rightarrow}{\omega}=\frac{{\Sigma^{-1}}^T\mathbf{1}}{\mathbf{1}^T\Sigma^{-1}\mathbf{1}}$$

$\Sigma^{-1}$ is symmetric matrix, $$\Rightarrow~ \stackrel{\rightarrow}{\omega}=\frac{\Sigma^{-1}\mathbf{1}}{\mathbf{1}^T\Sigma^{-1}\mathbf{1}}$$

$$ \stackrel{\rightarrow}{\omega}_{model} = \frac{\Sigma_{model}^{-1}\mathbf{1}}{\mathbf{1}^T\Sigma_{model}^{-1}\mathbf{1}}~,~~~ \stackrel{\rightarrow}{\omega}_{sample} = \frac{\Sigma_{sample}^{-1}\mathbf{1}}{\mathbf{1}^T\Sigma_{sample}^{-1}\mathbf{1}}$$
